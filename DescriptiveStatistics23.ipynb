```python
import json

with open("data/world_happiness_data.json", "r") as f:
    data = json.load(f)

life_ladder_23 = {}
for country in data:
    for year in data[country]:
        if year == "2023":
            life_ladder_23[country] = float(data[country][year]["Life Ladder"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the dictionary into a DataFrame
df_happiness = pd.DataFrame(list(life_ladder_23.items()), columns=['Country', 'Happiness_Score'])

# Display the first few entries of the DataFrame
print(df_happiness.head())

# Basic statistical summary of the DataFrame
print(df_happiness.describe())

# Plotting the distribution of happiness scores
plt.figure(figsize=(10, 6))
sns.histplot(df_happiness['Happiness_Score'], kde=True)
plt.title('Distribution of Happiness Scores for 2023')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

# Optional: Additional analyses such as identifying the highest and lowest scores
print("Countries with the highest Happiness Scores:")
print(df_happiness.nlargest(5, 'Happiness_Score'))

print("Countries with the lowest Happiness Scores:")
print(df_happiness.nsmallest(5, 'Happiness_Score'))
```
