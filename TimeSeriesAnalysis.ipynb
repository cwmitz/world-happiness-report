```python
import json

with open("data/life_ladder.json", "r") as f:
    data = json.load(f)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is already loaded from the JSON file as you have done.
# Convert JSON data to a DataFrame
country_data = []
for country, years in data.items():
    for year, happiness_score in years.items():
        country_data.append({"Country": country, "Year": int(year), "Happiness": happiness_score})

df = pd.DataFrame(country_data)

# Filter data for the years 2012 to 2023
df = df[df['Year'].between(2012, 2023)]

# Calculate the average happiness score for each year
yearly_averages = df.groupby('Year')['Happiness'].mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(yearly_averages.index, yearly_averages.values, marker='o')
plt.title('Average Happiness Score by Year (2005-2023)')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)
plt.show()
```
