```python
import json

with open("data/world_happiness_data.json", "r") as f:
    data = json.load(f)

data_23 = {}
for country in data:
    for year in data[country]:
        if year == "2023":
            data_23[country] = data[country][year]
```

```python
# Convert to pandas DataFrame
import pandas as pd

df = pd.DataFrame(data_23).T

# Compute correlation matrix
corr = df.corr()

# Plot correlation matrix
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True)
plt.show()

# Compute correlation with Life Ladder
corr_with_life_ladder = corr["Life Ladder"].sort_values(ascending=False)
print(corr_with_life_ladder)

# Plot correlation with Life Ladder
plt.figure(figsize=(10, 5))
corr_with_life_ladder.plot(kind="bar")
plt.show()
```
